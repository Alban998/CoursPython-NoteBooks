### Modulo
Le modulo nous permet d'obtenir le reste dans une division et utilise le symbole %. Par exemple:

```python
19 % 5
```

```python
6%2
```

Notez que si un nombre est entièrement divisible sans reste, le résultat de l'appel modulo est 0. Nous pouvons l'utiliser pour tester les nombres pairs, car si un nombre modulo 2 est égal à 0, cela signifie que c'est un nombre pair !

```python
#print even number
for n in list1:
    if n % 2 == 0:
        print(n)
```

```python
for n in list1:
    if n % 2 == 0:
        print(n)
    else:
        print('Odd number')
```

# break, continue, pass

Nous pouvons utiliser les instructions <code>break</code>, <code>continue</code> et <code>pass</code> dans nos boucles pour ajouter des fonctionnalités supplémentaires dans divers cas. Les trois énoncés sont définis par :

     break : sort de la boucle englobante la plus proche actuelle.
     continue : va au début de la boucle englobante la plus proche.
     pass : ne fait rien du tout.
    
    



```python
a = 0

while a < 10:
    print('a is currently: ',a)
    print(' a is still less than 10, adding 1 to a')
    a+=1
    if a==3:
        print('a==3')
    else:
        print('continuing...')
        continue
```

**Écrivez un programme qui affiche les nombres entiers de 1 à 100. Mais pour les multiples de trois, affichez "everyone" au lieu du nombre, et pour les multiples de cinq, afficher "world". Pour les nombres qui sont des multiples de trois et de cinq, écrivez "hello".**

```python
for num in range(1,101):
    if num % 3 == 0 and num % 5 == 0:
        print("hello")
    elif num % 3 == 0:
        print("everyone")
    elif num % 5 == 0:
        print("world")
    else:
        print(num)
```

**Utilisez <code>for</code>, .split() et <code>if</code> pour créer une déclaration qui affichera les mots commençant par 's' :**

```python
st = 'Print only the words that start with s in this sentence'
```

```python
for word in st.split():
    if word[0] == 's':
        print(word)
```
