```python
# Exercice 1 : Ecrire un algorithme qui affiche le plus petit entier parmi trois entiers lus au clavier.

def min_val(a, b, c):
    min_val = a
    if (b <= min_val):
        min_val = b
    if (c <= min_val):
        min_val = c
    return min_val

min_val(7, 5, 2)
```
