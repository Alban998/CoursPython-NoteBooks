
Identifiers

In Python, identifierest le nom donné aux entités comme la classe, les fonctions, les variables, etc. Il aide à différencier une entité d'une autre.

Règles d'écriture des Identifiants :

     Les identificateurs peuvent être une combinaison de lettres en minuscules (a-z) ou en majuscules (A-Z) ou de chiffres (0-9) ou de traits de soulignement (_).
     Les identifiants ne peuvent pas commencer par un chiffre. Par exemple : 12lose n'est pas valide, mais win12 est valide.
     Les mots-clés ne peuvent pas être utilisés comme identifiants.

## in operator

Nous avons déjà vu le mot-clé **in** lors de la boucle for, nous pouvons également l'utiliser pour vérifier rapidement si un objet est dans une liste.

```python
'x' in ['x','y','z']
```

```python
'a' in ['x','y','z']
```

## not in

Nous pouvons combiner **in** avec un opérateur **not**, pour vérifier si un objet ou une variable n'est pas présent dans une liste.

```python
'x' not in ['x','y','z']
```

```python
'a' not in ['x','y','z']
```

# Opérateurs de comparaison chaînés

Une fonctionnalité intéressante de Python est la possibilité de *chaîner* plusieurs comparaisons pour effectuer un test plus complexe. Vous pouvez utiliser ces comparaisons chaînées comme raccourci pour des expressions booléennes plus grandes.

```python
5 < 10 < 20
```

L'instruction ci-dessus vérifie si 5 était inférieur à 10 **and** si 10 était inférieur à 20. Nous aurions pu écrire ceci en utilisant une instruction **and** en Python :

```python
5<10 and 10<20
```

Il est important de noter que Python vérifie les deux instances des comparaisons. Nous pouvons également utiliser ** or ** pour écrire des comparaisons en Python. Par exemple:

```python
5==2 or 2<3
```

Notez comment c'était vrai; c'est parce qu'avec l'opérateur **or**, nous n'avons besoin que d'un *or* l'autre pour être vrai.

```python
12==12 or 100==50
```

Notez comment c'était vrai; c'est parce qu'avec l'opérateur **or**, nous n'avons besoin que d'un *ou* l'autre pour être vrai.
