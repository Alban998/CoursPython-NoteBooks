```python
# Compter le nombre de fois un nombre x se repete dans une liste :
liste = [12,23,12,3,4,5,3,6,4,3,3,2,13]

x = 3
compteur = 0

for i in liste:
    if i == x:
        compteur += 1

print("le nombre", x, "est répété :", compteur, "fois")
    
```

```python
# Compter le nombre de fois un nombre x se repete dans une liste (methode count):
liste = [12,23,12,3,4,5,3,6,4,3,3,2,13]

x = 3

print('le nombre',x,'est répété :',liste.count(x),'fois')
print(f'Le nombre {x} est repete {liste.count(x)} fois.') 
```

```python
# Compter le nombre de fois un nombre x se repete dans une liste (methode de fonction):
def nombreOccurences(L ,x):
    compteur = 0

    for i in L:
        if i == x:
            compteur += 1
    return compteur

L= [12,23,12,3,4,5,3,6,4,3,3,2,3]

print("le nombre d'occurences de 3 est : ",nombreOccurences(L,3))
```

```python
# Foction qui calcul la somme des elements d'une liste 
def Somme(L) :
    return sum(L)
L= [1,2,3]
print('la somme des elements de la liste L est : ',Somme(L))
```

```python
# Foction qui calcul la somme des elements d'une liste (methode classique)
def Somme(L) :
    S=0
    for i in L:
        S+= i
    return S
    
L= [1,2,3]
print('la somme des elements de la liste L est : ',Somme(L))
```

```python
# Foction qui calcul la multiplication des elements d'une liste 
def Multiplication(L) :
    M = 1
    for i in L:
        M*=i
    return M
L= [5,2,4]
print('la multiplication des elements de la liste L egale a  : ',Multiplication(L))
```
