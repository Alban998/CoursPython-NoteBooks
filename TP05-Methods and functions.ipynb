# Les fonctions



### Qu'est-ce qu'une fonction ?

Formellement, une fonction est un morceau de code utile qui regroupe un ensemble d'instructions afin qu'elles puissent être exécutées plusieurs fois. Ils peuvent également nous permettre de spécifier des paramètres pouvant servir d'entrées aux fonctions.

À un niveau plus fondamental, les fonctions nous permettent de ne pas avoir à écrire plusieurs fois le même code encore et encore. Par exemple, vérifier la longueur d'une séquence len() est une tâche courante, vous voudriez écrire une fonction qui peut le faire à plusieurs reprises à la commande.

##  fonction
* mot-clé def
* exemple simple d'une fonction
* appeler une fonction avec ()
* accepter les paramètres
* Affichage versus retour
* ajouter de la logique à l'intérieur d'une fonction
* plusieurs retours à l'intérieur d'une fonction
* ajout de boucles à l'intérieur d'une fonction


### def keyword

Voyons comment créer la syntaxe d'une fonction en Python. Il a la forme suivante :

```python
def name_of_function(arg1,arg2):
    '''
    This is where the function's Document String (docstring) goes.
    When you call help() on your function it will be printed out.
    '''
    # Do stuff here
    # Return desired result
```

On commence par <code>def</code> suivi d'un espace suivi du nom de la fonction. Essayez de garder les noms pertinents !

Vient ensuite une paire de parenthèses avec un certain nombre d'arguments séparés par une virgule.

Maintenant, voici l'étape importante, vous devez indenter pour commencer correctement le code à l'intérieur de votre fonction. Python utilise **whitespace** pour organiser le code.

Ensuite, vous verrez la docstring, c'est là que vous écrivez une description de base de la fonction. Les docstrings ne sont pas nécessaires pour les fonctions simple.

Après tout cela, vous commencez à écrire le code que vous souhaitez exécuter.



## Exemple simple d'une fonction

```python
def say_hello():
    print('Hello everyone!')
```

### Appel d'une fonction avec ()

```python
say_hello()
```

Si vous oubliez la parenthèse (), cela affichera simplement le fait que say_hello est une fonction. Plus tard, nous apprendrons que nous pouvons réellement passer des fonctions dans d'autres fonctions ! Mais pour l'instant, n'oubliez pas d'appeler les fonctions avec ().

```python
say_hello
```

### Accepter des paramètres (arguments)
Écrivons une fonction qui salue les gens avec leur nom.

```python
def greeting(name):
    print(f'Hello {name}')
```

```python
greeting("Amal")
```

## Utilisation du return

Jusqu'à présent, nous n'avons vu que print() utilisé, mais si nous voulons réellement enregistrer la variable résultante, nous devons utiliser le mot-clé return.



```python
def add_num(nb1,nb2):
    return nb1+nb2
```

```python
add_num(2,10)
```

##Question très courante : "Quelle est la différence entre *return* et *print* ?"

**Le mot-clé return vous permet d'enregistrer le résultat de la sortie d'une fonction en tant que variable. La fonction print() vous affiche simplement la sortie, mais ne l'enregistre pas pour une utilisation future. Explorons cela plus en détail **

```python
def print_result(x,y):
    print(x+y)
```

```python
def return_result(x,y):
    return x+y
```

```python
print_result(3,15)
```

```python
return_result(1,15)
```

**Mais que se passe-t-il si nous voulons réellement enregistrer ce résultat pour une utilisation ultérieure ?**

```python
my_result = print_result(3,2)
```

```python
type(my_result)
```

```python
my_result
```

```python
my_result = return_result(30,20)
```

```python
my_result
```

```python
type(my_result)
```

**Écrivez une fonction Python pour multiplier tous les nombres d'une liste.**

     Liste d'échantillons : [ 0.5, 10, 2]
     Sortie prévue : 10,0

```python
## Solution
def multiplication(NB):  
    tot = 1
    for i in NB:
        tot *= i
    return tot
```

```python
#call function
multiplication([0.5,10,2])
```

## EX 02:
____
**Écrivez une fonction Python qui prend une liste et renvoie une nouvelle liste avec des éléments uniques de la première liste.**

     Liste d'échantillons : [0,0,0,5,4,5,4,2,11,18,19,5,5,18,18,19]
     Liste Unique : [0,5,4,2,11,18,19]

```python
def uniq_list(lst):
    # Also possible to use list(set())
    x = []
    for e in lst:
        if e not in x:
            x.append(e)
    return x
```

```python
uniq_list([0,0,0,5,4,5,4,2,11,18,19,5,5,18,18,19])
```
