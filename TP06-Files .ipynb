### Writing to a File

Par défaut, la fonction `open()` nous permettra uniquement de lire le fichier. Nous devons passer l'argument `'w'` pour écrire sur le fichier. 

```python
myfile = open('test.txt','w+')
```

```python
myfile.read()
```

lire avec 'w' et 'w+' va **supprimer** tout le contenu qui a déjà été écrit dans le fichier d'origine

```python
# Write to the file
myfile.write('This is what will be written in the file')
```

```python
myfile.close()
```
