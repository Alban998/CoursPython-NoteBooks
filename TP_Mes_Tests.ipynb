```python
# Ecrire un programme python qui permet de chercher le mot le plus long sur une chaine g
g = input("Veuillez saisir une chaine de caracteres: ")
mots = g.split()
mot_le_plus_long = max(mots, key=len)
print(f"Le mot le plus long dans la chaine est: '{mot_le_plus_long}'")

# seconde version
g = input("Veuillez saisir une chaine de caracteres: ")
mots = g.split()
mot_le_plus_long = ""
for mot in mots:
    if len(mot) > len(mot_le_plus_long):
        mot_le_plus_long = mot
print(f"Le mot le plus long dans la chaine est: '{mot_le_plus_long}'")

# troisieme version
def maxMot(g):
    mots = g.split()
    mot_le_plus_long = ""
    for mot in mots:
        if len(mot) > len(mot_le_plus_long):
            mot_le_plus_long = mot
    return mot_le_plus_long

g = input("Veuillez saisir une chaine de caracteres: ")
mot_le_plus_long = maxMot(g)
print(f"Le mot le plus long dans la chaine est: '{mot_le_plus_long}'")
```

```python
# Creez une fonction Python, appelee nombreDivisibles, qui s'applique a une liste de nombres et un entier n, et qui renvoie le nombre d'elements de la liste qui sont divisibles par n.
def nombreDivisibles(liste_nombres, n):
    cpt = 0
    for nombre in liste_nombres:
        if nombre % n == 0:
            cpt += 1
    return cpt  

# Exemple d'utilisation de la fonction nombreDivisibles 
input_liste = input("Veuillez saisir une liste de nombres entiers separes par des espaces: ")
liste_nombres = list(map(int, input_liste.split()))
n = int(input("Veuillez saisir un entier n: "))
resultat = nombreDivisibles(liste_nombres, n)
print(f"Le nombre d'elements de la liste divisibles par {n} est: {resultat}")

```

```python
# implementation de la descente de gradient pour minimiser la fonction f(x) = x^2 + y^2 + sin(x+y)
import torch

v = torch.tensor([1.0, 1.0], requires_grad=True)

def f_torh(y) :
    x, y = v
    return x**2 + y**2 + torch.sin(x + y)

def grad_f(x, y) :
    # requires_grad active le suivi des operations sur le tenseur
    v = torch.tensor([x, y], requires_grad=True)
    # Evaluation de la fonction
    val = f_torh(v)
    # Calcul du gradient
    val.backward()
    return v.grad.numpy()
```

```python
# Ecrivez uen fonction Python qui prend une liste et renvoie une nouvelle liste avec les elements uniques de la liste d'origine (c'est-a-dire sans doublons).

def elements_uniques(liste):
    liste_unique = []
    for element in liste:
        if element not in liste_unique:
            liste_unique.append(element)
    return liste_unique
```

```python
# Ecrire un programme Python qui permet de calculer la somme des elements d'une liste donnee
def somme_elements(liste):
    somme = 0
    for element in liste:
        somme += element
    return somme

def moyenne_elements(liste):
    if len(liste) == 0:
        return 0
    somme = somme_elements(liste)
    moyenne = somme / len(liste)
    return moyenne

# Exemple d'utilisation de la fonction somme_elements
input_liste = input("Veuillez saisir une liste de nombres entiers separes par des espaces: ")
liste_nombres = list(map(int, input_liste.split()))
resultat = somme_elements(liste_nombres)

print(f"La somme des elements de la liste est: {resultat}")
```

```python
# Ecrire une fonction qui prend une liste de nombres et determine si tous sont differents les uns des autres.
def tous_differents(liste_nombres):
    ensemble_nombres = set()
    for nombre in liste_nombres:
        if nombre in ensemble_nombres:
            return False
        ensemble_nombres.add(nombre)
    return True
```
